```python
# Import libraries for text processing and clustering
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import warnings
warnings.filterwarnings('ignore')

# Download required NLTK data
try:
    nltk.data.find('tokenizers/punkt')
    nltk.data.find('corpora/stopwords')
    nltk.data.find('corpora/wordnet')
except LookupError:
    nltk.download('punkt')
    nltk.download('stopwords')
    nltk.download('wordnet')

print("Text processing libraries loaded successfully!")
```

```python
# Text preprocessing function for abstracts
def preprocess_abstract(text):
    """
    Advanced preprocessing for research abstracts
    """
    if pd.isna(text) or not isinstance(text, str):
        return ""
    
    # Convert to lowercase and remove special characters, but keep scientific terms
    text = re.sub(r'[^\w\s]', ' ', text.lower())
    
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Ensure punkt_tab is downloaded
    import nltk
    try:
        nltk.data.find('tokenizers/punkt_tab')
    except LookupError:
        nltk.download('punkt_tab')
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stopwords but keep scientific terms
    stop_words = set(stopwords.words('english'))
    # Remove common non-informative words but keep domain-specific terms
    additional_stops = {'would', 'could', 'should', 'may', 'might', 'project', 'research', 'study', 'investigation'}
    stop_words.update(additional_stops)
    
    # Filter tokens: remove stopwords, very short words, and numbers
    tokens = [token for token in tokens if 
             len(token) > 2 and 
             token not in stop_words and 
             not token.isdigit()]
    
    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    return ' '.join(tokens)

# Clean the data first
print("Cleaning and preprocessing data...")
data_clean = data_nih.dropna(subset=['abstract_text', 'direct_cost']).copy()
print(f"Original data shape: {data_nih.shape}")
print(f"Clean data shape: {data_clean.shape}")

# Preprocess abstracts
print("Preprocessing abstracts...")
data_clean['processed_abstract'] = data_clean['abstract_text'].apply(preprocess_abstract)

# # Remove rows with empty processed abstracts
# data_clean = data_clean[data_clean['processed_abstract'].str.len() > 10]
# print(f"Final clean data shape: {data_clean.shape}")

# # Display sample of processed abstract
# print(f"\nSample processed abstract:")
# print(data_clean['processed_abstract'].iloc[0][:500] + "...")
```

```python
# Create TF-IDF encoder for text vectorization
print("Creating TF-IDF vectorizer...")
vectorizer = TfidfVectorizer(
    max_features=500,      # Top 500 most important features
    min_df=3,              # Ignore terms appearing in fewer than 3 documents
    max_df=0.7,            # Ignore terms appearing in more than 70% of documents
    ngram_range=(1, 2),    # Use both single words and bigrams
    stop_words='english'   # Additional stopword filtering
)

# Fit and transform the abstracts
print("Encoding abstracts using TF-IDF...")
tfidf_matrix = vectorizer.fit_transform(data_clean['processed_abstract'])
feature_names = vectorizer.get_feature_names_out()

print(f"TF-IDF matrix shape: {tfidf_matrix.shape}")
print(f"Number of features (terms): {len(feature_names)}")
print(f"Sample features: {list(feature_names[:20])}")
```

```python
# Apply K-Means clustering to categorize research areas
n_clusters = 8  # Number of research categories to identify

print(f"Performing K-Means clustering with {n_clusters} clusters...")
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(tfidf_matrix)

# Add cluster labels to the dataframe
data_clean['research_cluster'] = cluster_labels
data_clean['cluster_id'] = cluster_labels

print("Clustering complete!")
print(f"Cluster distribution:")
cluster_counts = pd.Series(cluster_labels).value_counts().sort_index()
print(cluster_counts)

# Calculate cluster centers to understand what each cluster represents
cluster_centers = kmeans.cluster_centers_

print(f"\nCluster centers shape: {cluster_centers.shape}")
```

```python
# Analyze and label each cluster based on top keywords
def get_cluster_keywords(cluster_id, n_keywords=10):
    """
    Get top keywords that define each cluster
    """
    cluster_center = cluster_centers[cluster_id]
    
    # Get indices of top features forna this cluster
    top_indices = cluster_center.argsort()[-n_keywords:][::-1]
    
    # Get corresponding feature names and scores
    top_keywords = [(feature_names[i], cluster_center[i]) for i in top_indices]
    
    return top_keywords

# Analyze each cluster
print("RESEARCH AREA CLUSTER ANALYSIS")
print("=" * 60)

cluster_analysis = {}
for cluster_id in range(n_clusters):
    keywords = get_cluster_keywords(cluster_id)
    cluster_size = sum(cluster_labels == cluster_id)
    
    print(f"\nCLUSTER {cluster_id}: ({cluster_size} projects)")
    print("-" * 30)
    print("Top Keywords:")
    for keyword, score in keywords[:8]:
        print(f"   {keyword}: {score:.4f}")
    
    cluster_analysis[cluster_id] = {
        'keywords': [kw[0] for kw in keywords[:5]],
        'size': cluster_size,
        'top_keyword_scores': keywords
    }

print("\n" + "=" * 60)
```

```python
# Assign meaningful names to research areas based on clustering results
# You can modify these names based on the actual keywords that appear in your analysis

research_area_names = {
    0: "Biomedical & Health Sciences",
    1: "Engineering & Technology", 
    2: "Environmental & Earth Sciences",
    3: "Computer Science & Data Analysis",
    4: "Materials Science & Chemistry",
    5: "Education & Social Sciences",
    6: "Physics & Mathematical Sciences",
    7: "Life Sciences & Biology"
}

# Map cluster IDs to research area names
data_clean['research_area'] = data_clean['research_cluster'].map(research_area_names)

print("Research areas assigned:")
print(data_clean['research_area'].value_counts())
```

```python
# Automatically generate research area names by joining top keywords from each cluster

def auto_name_research_area(keywords, n=3):
    """
    Assign a research area name by joining the top N keywords.
    """
    return " / ".join(keywords[:n])

# Generate research area names for each cluster
if 'n_clusters' not in globals():
    n_clusters = len(cluster_analysis)

auto_area_names = {}
for cluster_id in range(n_clusters):
    top_keywords = cluster_analysis[cluster_id]['keywords']
    auto_area_names[cluster_id] = auto_name_research_area(top_keywords)

# Map cluster IDs to auto-generated research area names
data_clean['research_area'] = data_clean['research_cluster'].map(auto_area_names)

print("Auto-generated research areas based on cluster keywords:")
for cid, name in auto_area_names.items():
    print(f"Cluster {cid}: {name} (keywords: {', '.join(cluster_analysis[cid]['keywords'])})")

print("\nResearch area assignment complete.")
print(data_clean['research_area'].value_counts())
```

```python
# Import BERT-related libraries
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("BERT libraries imported successfully!")
```

```python
def preprocess_text_for_bert(text):
    """
    Enhanced preprocessing specifically for BERT embeddings.
    Removes stop words while preserving semantic content.
    """
    if pd.isna(text) or not isinstance(text, str):
        return ""
    
    # Convert to lowercase and clean formatting
    text = text.lower()
    text = re.sub(r'\r\n|\n|\r', ' ', text)  # Remove line breaks
    text = re.sub(r'[^\w\s]', ' ', text)     # Remove punctuation but keep spaces
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra whitespace
    
    # Download NLTK data if needed
    try:
        nltk.data.find('tokenizers/punkt_tab')
        nltk.data.find('corpora/stopwords')
        nltk.data.find('corpora/wordnet')
    except LookupError:
        nltk.download('punkt_tab')
        nltk.download('stopwords')
        nltk.download('wordnet')
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Define comprehensive stop words
    stop_words = set(stopwords.words('english'))
    
    # Add common research paper stop words
    additional_stops = {
        'will', 'would', 'could', 'should', 'may', 'might', 'must', 'shall',
        'can', 'cannot', 'able', 'unable', 'also', 'however', 'therefore',
        'thus', 'hence', 'furthermore', 'moreover', 'additionally', 'meanwhile',
        'nevertheless', 'nonetheless', 'although', 'though', 'whereas', 'while',
        'since', 'because', 'due', 'according', 'regarding', 'concerning',
        'including', 'excluding', 'following', 'during', 'within', 'without',
        'among', 'between', 'through', 'across', 'around', 'under', 'over',
        'project', 'study', 'research', 'investigation', 'analysis', 'program',
        'grant', 'funding', 'support', 'award', 'application', 'proposal',
        'description', 'abstract', 'summary', 'background', 'objective',
        'purpose', 'goal', 'aim', 'method', 'approach', 'technique', 'procedure',
        'result', 'finding', 'outcome', 'conclusion', 'discussion', 'future',
        'potential', 'possible', 'likely', 'unlikely', 'significant', 'important',
        'major', 'minor', 'primary', 'secondary', 'main', 'key', 'critical',
        'essential', 'necessary', 'required', 'needed', 'used', 'using',
        'provide', 'provided', 'providing', 'develop', 'developed', 'developing',
        'determine', 'determined', 'determining', 'identify', 'identified',
        'identifying', 'examine', 'examined', 'examining', 'investigate',
        'investigated', 'investigating', 'evaluate', 'evaluated', 'evaluating',
        'assess', 'assessed', 'assessing', 'analyze', 'analyzed', 'analyzing',
        'compare', 'compared', 'comparing', 'demonstrate', 'demonstrated',
        'demonstrating', 'show', 'shown', 'showing', 'indicate', 'indicated',
        'indicating', 'suggest', 'suggested', 'suggesting', 'reveal', 'revealed',
        'revealing', 'establish', 'established', 'establishing'
    }
    
    stop_words.update(additional_stops)
    
    # Filter tokens: remove stop words, very short words, and numbers
    filtered_tokens = []
    for token in tokens:
        if (len(token) > 2 and 
            token not in stop_words and 
            not token.isdigit() and
            not token.startswith('http') and
            not re.match(r'^[0-9]+$', token)):
            filtered_tokens.append(token)
    
    # Lemmatize to reduce word variations
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    
    return ' '.join(lemmatized_tokens)

def create_bert_embeddings(texts, model_name='all-MiniLM-L6-v2'):
    """
    Create BERT embeddings for research abstracts using SentenceTransformers.
    
    Parameters:
    - texts: List of text strings (research abstracts)
    - model_name: Pre-trained model to use for embeddings
    
    Returns:
    - embeddings: NumPy array of shape (n_texts, embedding_dim)
    """
    print(f"Loading BERT model: {model_name}")
    
    # Initialize the sentence transformer model
    model = SentenceTransformer(model_name)
    
    print(f"Creating embeddings for {len(texts)} texts...")
    
    # Generate embeddings
    embeddings = model.encode(texts, 
                             show_progress_bar=True,
                             convert_to_numpy=True,
                             normalize_embeddings=True)
    
    print(f"Embeddings created successfully! Shape: {embeddings.shape}")
    return embeddings, model

def perform_semantic_clustering(embeddings, n_clusters=8, random_state=42):
    """
    Perform K-means clustering on BERT embeddings for semantic grouping.
    
    Parameters:
    - embeddings: BERT embeddings array
    - n_clusters: Number of clusters to create
    - random_state: Random state for reproducibility
    
    Returns:
    - cluster_labels: Array of cluster assignments
    - kmeans_model: Fitted KMeans model
    """
    print(f"Performing semantic clustering with {n_clusters} clusters...")
    
    # Apply K-means clustering
    kmeans = KMeans(n_clusters=n_clusters, 
                   random_state=random_state, 
                   n_init=20,
                   max_iter=300)
    
    cluster_labels = kmeans.fit_predict(embeddings)
    
    print("Semantic clustering completed!")
    print(f"Cluster distribution:")
    unique, counts = np.unique(cluster_labels, return_counts=True)
    for cluster_id, count in zip(unique, counts):
        print(f"  Cluster {cluster_id}: {count} projects")
    
    return cluster_labels, kmeans

def analyze_semantic_clusters(data, cluster_labels, text_column='processed_abstract', n_clusters=8):
    """
    Analyze semantic clusters by extracting representative terms and examples.
    
    Parameters:
    - data: DataFrame with research data
    - cluster_labels: Array of cluster assignments
    - text_column: Column name containing processed text
    - n_clusters: Number of clusters
    
    Returns:
    - cluster_analysis: Dictionary with cluster information
    """
    print("Analyzing semantic clusters...")
    
    # Add cluster labels to dataframe
    data_with_clusters = data.copy()
    data_with_clusters['semantic_cluster'] = cluster_labels
    
    cluster_analysis = {}
    
    for cluster_id in range(n_clusters):
        # Get texts in this cluster
        cluster_mask = cluster_labels == cluster_id
        cluster_texts = data_with_clusters[cluster_mask][text_column].tolist()
        cluster_projects = data_with_clusters[cluster_mask]
        
        # Extract common terms (using processed clean text)
        all_words = []
        for text in cluster_texts:
            if isinstance(text, str):
                # Apply the same preprocessing to extract clean words
                clean_text = preprocess_text_for_bert(text)
                all_words.extend(clean_text.split())
        
        # Count word frequencies and get meaningful terms
        from collections import Counter
        word_freq = Counter(all_words)
        # Filter for meaningful terms (length > 3, not purely numeric)
        meaningful_words = [word for word, count in word_freq.most_common(15) 
                           if len(word) > 3 and not word.isdigit() and count > 1]
        
        # Get sample project titles
        sample_titles = cluster_projects['project_title'].head(3).tolist()
        
        cluster_analysis[cluster_id] = {
            'size': sum(cluster_mask),
            'top_words': meaningful_words[:10],
            'sample_titles': sample_titles,
            'avg_funding': cluster_projects['award_amount'].mean(),
            'total_funding': cluster_projects['award_amount'].sum()
        }
        
        print(f"\nSemantic Cluster {cluster_id}: ({sum(cluster_mask)} projects)")
        print(f"  Key terms: {', '.join(meaningful_words[:6])}")
        print(f"  Avg funding: ${cluster_analysis[cluster_id]['avg_funding']:,.2f}")
        print(f"  Sample project: {sample_titles[0][:80]}..." if sample_titles else "  No projects")
    
    return cluster_analysis, data_with_clusters

print("Enhanced BERT-based semantic clustering functions created successfully!")
print("Now includes comprehensive stop word removal and better preprocessing!")
```

```python
# Apply BERT-based semantic clustering to research abstracts with enhanced preprocessing
print("=" * 60)
print("BERT-BASED SEMANTIC CLUSTERING ANALYSIS")
print("(WITH ENHANCED STOP WORD REMOVAL)")
print("=" * 60)

# Prepare clean abstracts for BERT processing using enhanced preprocessing
clean_abstracts = []
valid_indices = []

print("Applying enhanced preprocessing with comprehensive stop word removal...")

for idx, abstract in enumerate(data_clean['abstract_text']):
    if pd.notna(abstract) and isinstance(abstract, str) and len(abstract.strip()) > 50:
        # Apply enhanced preprocessing that removes stop words
        clean_text = preprocess_text_for_bert(abstract)
        
        # Only include if we have meaningful content after preprocessing
        if len(clean_text.strip()) > 20:  # Ensure sufficient content remains
            clean_abstracts.append(clean_text)
            valid_indices.append(idx)

print(f"Processing {len(clean_abstracts)} abstracts with BERT...")
print(f"Sample cleaned text: {clean_abstracts[0][:200]}...")

# Create BERT embeddings using cleaned text
bert_embeddings, bert_model = create_bert_embeddings(clean_abstracts)

# Filter the dataframe to match the valid abstracts
data_bert = data_clean.iloc[valid_indices].copy().reset_index(drop=True)

print(f"Data shape after filtering: {data_bert.shape}")
print(f"BERT embeddings shape: {bert_embeddings.shape}")
```

```python
# Create semantic research area names based on cluster analysis
def generate_semantic_area_names(semantic_analysis):
    """
    Generate meaningful research area names based on semantic clustering results.
    """
    area_names = {}
    
    for cluster_id, info in semantic_analysis.items():
        # Use top words and sample titles to create meaningful names
        top_words = info['top_words'][:3]
        
        # Create area name from top words
        if len(top_words) >= 2:
            area_name = f"{top_words[0].title()} & {top_words[1].title()}"
        elif len(top_words) == 1:
            area_name = f"{top_words[0].title()} Research"
        else:
            area_name = f"Research Area {cluster_id + 1}"
            
        area_names[cluster_id] = area_name
    
    return area_names

# Generate semantic area names
semantic_area_names = generate_semantic_area_names(semantic_analysis)

# Add semantic research areas to dataframe
data_with_semantic['semantic_research_area'] = data_with_semantic['semantic_cluster'].map(semantic_area_names)

print("\\nSemantic Research Areas Generated:")
print("-" * 40)
for cluster_id, name in semantic_area_names.items():
    size = semantic_analysis[cluster_id]['size']
    funding = semantic_analysis[cluster_id]['total_funding']
    print(f"Cluster {cluster_id}: {name} ({size} projects, ${funding:,.2f})")

# Calculate funding analysis for semantic clusters
semantic_funding_analysis = data_with_semantic.groupby('semantic_research_area').agg({
    'award_amount': ['sum', 'mean', 'count', 'median']
}).round(2)

semantic_funding_analysis.columns = ['Total_Funding', 'Avg_Funding', 'Project_Count', 'Median_Funding']
semantic_funding_analysis = semantic_funding_analysis.sort_values('Total_Funding', ascending=False)
semantic_funding_analysis['Funding_Percentage'] = (semantic_funding_analysis['Total_Funding'] / semantic_funding_analysis['Total_Funding'].sum() * 100).round(2)

print(f"\\n\\nSEMANTIC CLUSTERING - FUNDING ANALYSIS")
print("=" * 50)
print(semantic_funding_analysis)
```

```python
# Compare TF-IDF vs BERT clustering results
def compare_clustering_methods(data_clean, data_with_semantic):
    """
    Compare the clustering results between TF-IDF and BERT methods.
    """
    print("\\n" + "=" * 60)
    print("COMPARISON: TF-IDF vs BERT CLUSTERING")
    print("=" * 60)
    
    # Align datasets for comparison (both should have same number of projects)
    # Get original TF-IDF funding analysis
    tfidf_funding = data_clean.groupby('research_area').agg({
        'award_amount': ['sum', 'mean', 'count']
    }).round(2)
    tfidf_funding.columns = ['Total_Funding', 'Avg_Funding', 'Project_Count']
    tfidf_funding = tfidf_funding.sort_values('Total_Funding', ascending=False)
    
    print("\\nTOP 5 RESEARCH AREAS - TF-IDF METHOD:")
    print("-" * 45)
    for i, (area, row) in enumerate(tfidf_funding.head(5).iterrows(), 1):
        print(f"{i}. {area}")
        print(f"   Total: ${row['Total_Funding']:,.2f} | Projects: {row['Project_Count']}")
    
    print("\\nTOP 5 RESEARCH AREAS - BERT METHOD:")
    print("-" * 45)
    for i, (area, row) in enumerate(semantic_funding_analysis.head(5).iterrows(), 1):
        print(f"{i}. {area}")
        print(f"   Total: ${row['Total_Funding']:,.2f} | Projects: {row['Project_Count']}")
    
    # Calculate clustering quality metrics
    from sklearn.metrics import silhouette_score, calinski_harabasz_score
    
    print("\\n\\nCLUSTERING QUALITY METRICS:")
    print("-" * 35)
    
    # For TF-IDF clustering
    if 'tfidf_matrix' in globals():
        tfidf_silhouette = silhouette_score(tfidf_matrix.toarray(), data_clean['research_cluster'])
        tfidf_calinski = calinski_harabasz_score(tfidf_matrix.toarray(), data_clean['research_cluster'])
        print(f"TF-IDF Method:")
        print(f"  Silhouette Score: {tfidf_silhouette:.4f}")
        print(f"  Calinski-Harabasz Score: {tfidf_calinski:.2f}")
    
    # For BERT clustering  
    bert_silhouette = silhouette_score(bert_embeddings, semantic_labels)
    bert_calinski = calinski_harabasz_score(bert_embeddings, semantic_labels)
    print(f"\\nBERT Method:")
    print(f"  Silhouette Score: {bert_silhouette:.4f}")
    print(f"  Calinski-Harabasz Score: {bert_calinski:.2f}")
    
    print("\\nInterpretation:")
    print(" Higher Silhouette Score (closer to 1) = better defined clusters")
    print(" Higher Calinski-Harabasz Score = better separated clusters")
    
    return tfidf_funding, semantic_funding_analysis

# Perform the comparison
tfidf_results, bert_results = compare_clustering_methods(data_clean, data_with_semantic)
```
